```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Get the last 12 months' actual values (matching LSTM forecast period)
y_test_actual = scaler.inverse_transform(df_scaled[-12:].reshape(-1,1))

# Ensure both arrays have the same shape
print("y_test_actual shape:", y_test_actual.shape)
print("lstm_forecast shape:", lstm_forecast.shape)

# Evaluate the model
mae = mean_absolute_error(y_test_actual, lstm_forecast)
mse = mean_squared_error(y_test_actual, lstm_forecast)
r2 = r2_score(y_test_actual, lstm_forecast)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R² Score: {r2}")
```
